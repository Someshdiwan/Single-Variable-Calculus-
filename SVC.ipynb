```python
from sympy import Symbol, diff, solve

# Define the variable and function
x = Symbol('x')
f = x**3 + 4*(x**2) + 5*x

# First derivative of the function
f1 = diff(f, x)

# Finding the roots of the first derivative
c = solve(f1, x)
print("ROOTS AFTER DERIVATIVES:", c)

# Define intervals and evaluate function values
a = 0
b = 5

c1 = f.subs(x, a)
print("VALUE AT a:", c1)

c2 = f.subs(x, b)
print("VALUE AT b:", c2)


for point in c:
    c3 = f.subs(x, point)
    print(f"VALUE AT critical point {point}:", c3)

c4 = f1.subs(x, a)
print("VALUE OF DERIVATIVE AT a:", c4)
```

```python
from sympy import *
x=symbols('x')
y=Function('y')(x)
print('The given differential eq. is:')
diffeq=Eq(y.diff(x,x,x)-y.diff(x,x)-18*y,0)
display(diffeq)
D = symbols('D')
print('Auxiliary eq. is:')
ae=Eq(D**3 -D**2 -18, 0)
display(ae)
roots=solve(D**3 -D**2 -18, D)
print('The roots of auxiliary eq. are:')
display(roots)
print('The required solutionof given differential eq. is:')
dsolve(diffeq,y)
print('Using given initial conditions we get')
dsolve(diffeq,y,ics={y.subs(x,0):2, y.diff(x).subs(x,0):3})
```
